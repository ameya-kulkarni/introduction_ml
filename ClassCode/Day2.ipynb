```python
class MLP():
    def __init__(self,input_dim):
        self.input_dim=input_dim
        self.layers=[]
    def _add_layer(self,input_shape,n,act):
        W=np.random.randn(input_shape,n)
        b=np.random.randn(n)
        act=act
        if act not in ["sigmoid","softmax"]:
            raise Exception("Activation function should be either sigmoid or softmax")
        params={'W':W,'b':b,'activation':act}
        return params
    def add_layer(self,n,act):
        if len(self.layers)==0:
            p=self._add_layer(input_shape=self.input_dim,n=n,act=act)
            self.layers.append(p)
        else:
            input_dim=self.layers[-1]['W'].shape[1]
            p=self._add_layer(input_shape=input_dim,n=n,act=act)
            self.layers.append(p)
    def sigmoid(self,z):
        activation=1/(1+np.exp(-z))
        return activation
    def softmax(self,z): #http://cs231n.github.io/linear-classify (Numerical Stability)
        e_z=np.exp(z-np.max(z))
        activation=e_z/e_z.sum(axis=1,keepdims=True)
        return activation
    def sigma_prime(self,z):
        sig=self.sigmoid(z)
        return np.multiply(sig,(1-sig))
    def forward(self,batch):
        self.activations=[]
        self.z=[]
        act_fn={'sigmoid':self.sigmoid,'softmax':self.softmax}
        for l in self.layers:
            batch=np.matmul(batch,l['W'])+l['b']
            self.z.append(batch)
            fn_name=l['activation']
            batch=act_fn[fn_name](batch)
            self.activations.append(batch)
        return batch
    def backward(self,y,batch):
        DW=[0]*len(self.layers)
        DB=[0]*len(self.layers)
        DELTA=[0]*len(self.layers)
        p=self.activations[-1]
        rows=len(y)
        cols=p.shape[1]
        y_act=np.zeros((rows,cols))
        y_act[range(rows),y]=1
        delta=p-y_act
        delta=delta/rows
        db=delta.sum(axis=0)
        dw=np.matmul(self.activations[-2].T,delta)
        DELTA[-1]=delta
        DW[-1]=dw
        DB[-1]=db
        for i in range(len(self.layers)-2,-1,-1):
            delta_prev=DELTA[i+1]
            wt_prev=self.layers[i+1]['W']
            act_prev=self.activations[i-1]
            z=self.z[i]
            delta=np.multiply(np.matmul(delta_prev,wt_prev.T),self.sigma_prime(z))
            DELTA[i]=delta/rows
            db=delta.sum(axis=0)
            DB[i]=db
            if i==0:
                dw=np.matmul(batch.T,delta)
            else:
                dw=np.matmul(act_prev.T,delta)
            DW[i]=dw
        self.DW=DW
        self.DB=DB
    def sgd_update(self,lr):
        for i in range(len(self.layers)):
            self.layers[i]['W']-=lr*self.DW[i]
            self.layers[i]['b']-=lr*self.DB[i]
```
